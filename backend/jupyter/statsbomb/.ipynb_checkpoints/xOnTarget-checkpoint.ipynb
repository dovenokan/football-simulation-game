```python
master_events_df = pd.read_csv("data/Bayer Leverkusen Bundesliga 23-24 Statsbomb Events Data.csv", low_memory=False)
# master_events_df = pd.read_csv("./data/master_events_processed.csv", low_memory=False)
# master_events_df = pd.read_csv("./data/Ligue1 22-23 Statsbomb Events Data.csv", low_memory=False)
```

```python
def safe_literal_eval(val):
    try:
        return ast.literal_eval(val) if isinstance(val, str) and val != '[]' else []
    except (ValueError, SyntaxError):
        return []

master_events_df['location'] = master_events_df['location'].fillna('[]')
master_events_df['location'] = master_events_df['location'].apply(safe_literal_eval)
master_events_df['y'] = master_events_df['location'].apply(lambda loc: loc[0] if loc else None)
master_events_df['x'] = master_events_df['location'].apply(lambda loc: loc[1] if loc else None)
```

```python
# x on target
master_events_df.iloc[3574].to_dict()
```

```python
shot_df = master_events_df.copy().query("type == 'Shot'")
```

```python
shot_df = shot_df.replace(['Saved to Post', 'Saved Off Target', 'Wayward' , 'Post'], 'Off T')
shot_df = shot_df.replace(['Saved'], 'Goal')
shot_df = shot_df.replace(['Goal'], 'On T')
shot_df["y"] = shot_df["y"].fillna(0).apply(lambda y: min(24, int(y * 24 / 120)))  
shot_df["x"] = shot_df["x"].fillna(0).apply(lambda x: min(16, int(x * 16 / 80)))   
shot_df["shot_on_target"] = shot_df["shot_outcome"].apply(lambda x: 1 if x == "On T" else 0)
```

```python
shot_df.shot_outcome.value_counts()
```

```python
shot_df.query("type == 'Shot'")[["type","y","x","shot_outcome", "shot_on_target"]]
```

```python
df = shot_df.copy()

# Initialize a pitch
pitch = Pitch(pitch_type='statsbomb', pitch_color='grass', line_color='white', goal_type='box')
fig, ax = pitch.draw(figsize=(12, 8))

# Define colors for different shot outcomes
outcome_colors = {
    'On T': 'red',
    'Blocked': 'black',
    'Off T': 'white'
}

# Plot the shots
for _, row in df.iterrows():
    ax.scatter(
        row['y'], row['x'], 
        color=outcome_colors[row['shot_outcome']], 
        edgecolors='black', s=100, label=row['shot_outcome'] if row['shot_outcome'] not in ax.get_legend_handles_labels()[1] else ""
    )

# Add legend
handles, labels = ax.get_legend_handles_labels()
by_label = dict(zip(labels, handles))
ax.legend(by_label.values(), by_label.keys(), loc='upper right', title="Shot Outcome")

# Add title
plt.title("Visualization of Shots", fontsize=15)
plt.show()
```

```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
import numpy as np

df = shot_df.copy() 

# Features (x, y) and target (on_target)
X = df[['x', 'y']]
y = df['shot_on_target']

# Train logistic regression
model = LogisticRegression()
model.fit(X, y)

# Model coefficients
beta_0 = model.intercept_[0]
beta_1, beta_2 = model.coef_[0]

print(f"Intercept (Œ≤‚ÇÄ): {beta_0}")
print(f"Coefficient for x (Œ≤‚ÇÅ): {beta_1}")
print(f"Coefficient for y (Œ≤‚ÇÇ): {beta_2}")

# Function to calculate P(On T)
def expected_on_target(y, x):
    if y < 13:
        return 0
    z = beta_0 + beta_1 * x + beta_2 * y
    return 1 / (1 + np.exp(-z))

# Example prediction
for y in range(1,24+1):
    for x in range(1,16+1):
        shot_location = (y, x)
        p_on_target = expected_on_target(*shot_location)
        print(f"Expected probability of On T for location {shot_location}: {p_on_target:.2f}")
```
