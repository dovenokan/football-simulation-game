# Hands-On StatsbombðŸ’Ž

```python
def load(file_id):
    df = pd.read_csv(f"data/all-events/split_part_{file_id}.csv", low_memory=False)[commoncols]
    return df
```

```python
def coordinator(df):
    def safe_literal_eval(val):
        try:
            return ast.literal_eval(val) if isinstance(val, str) and val != '[]' else []
        except (ValueError, SyntaxError):
            return []
    
    df['location'] = df['location'].fillna('[]')
    df['location'] = df['location'].apply(safe_literal_eval)
    df['y_yards'] = df['location'].apply(lambda loc: loc[0] if loc else None)
    df['x_yards'] = df['location'].apply(lambda loc: loc[1] if loc else None)
    df["y"] = df["y_yards"].fillna(0).apply(lambda y: min(24, int(y * 24 / 120)))  
    df["x"] = df["x_yards"].fillna(0).apply(lambda x: min(16, int(x * 16 / 80)))
    return df
```

```python
events_df_1 = load(1)
events_df_1 = coordinator(events_df_1)
```

```python
events_df_leverkusen = master_events_df.copy().sort_values(["match_id","timestamp"])
events_df_leverkusen = coordinator(events_df_leverkusen)
```

```python
df = pd.merge(events_df_1, matches_df, on="match_id", how="inner")
df = coordinator(df)
```

```python
import json

ignore_columns = ['location', 'id', 'player', 'team', 'angle', 'length', 'recipient', 'frame', 'xg']
ignored_columns = df.columns[df.columns.str.contains('|'.join(ignore_columns))]

# Create a dictionary to store the results
results = {}

# Copy the dataframe
events_df = df.copy()

# Remove the ignored columns
events_df = events_df.loc[:, ~events_df.columns.isin(ignored_columns)]

# Define actions to check
actions = [
    "ball_receipt",
    "ball_recovery",
    "block",
    "carry",
    "clearance",
    "dribble",
    "duel",
    "foul",
    "goalkeeper",
    "interception",
    "pass",
    "shot",
    "50_50",
    "counterpress",
    "miscontrol",
]

# Action-Based Columns Stats
action_stats = {}
for action in actions:
    relevant_columns = events_df[events_df.columns[events_df.columns.str.startswith(action)]].columns
    action_stats[action] = {}
    for col in relevant_columns:
        # Count valid (non-null) rows for each column
        valid_count = events_df[col].notna().sum()
        value_counts = events_df[col].value_counts().to_dict()
        action_stats[action][col] = {
            "valid_rows": int(valid_count),  # Convert to standard Python int
            "value_counts": value_counts
        }

results["Action-Based Columns Stats"] = action_stats

# Save the results to a JSON file
with open("data/all-events/event_stats.json", "w", encoding="utf-8") as f:
    json.dump(results, f, ensure_ascii=False, indent=4)
```

# Sequence, Sequence!

```python
# filtered_df = events_df[~events_df['type'].isin(["Tactical Shift","Half Start","Half End","Substitution",
#                                                  "Dribbled Past","Miscontrol","Foul Won","Referee Ball-Drop",
#                                                  "Injury Stoppage","","",""
#                                                 ])]

# events_df_1
filtered_df = events_df_leverkusen[~events_df_leverkusen['type'].isin(["Tactical Shift","Half Start","Half End","Substitution",
                                                 "Dribbled Past","Miscontrol","Foul Won","Referee Ball-Drop",
                                                 "Injury Stoppage","Error","Own Goal For","Tactical Shift",
                                                 "Starting XI","Player On","Shield","Player Off","Own Goal Against","Offside",
                                                 "Bad Behaviour","50/50","Ball Receipt*"
                                                ])].sort_values(["match_id","timestamp"])

filtered_df.loc[filtered_df['team'] == 'Bayer Leverkusen', 'type'] = 'ðŸ”´' + \
filtered_df.loc[filtered_df['team'] == 'Bayer Leverkusen', 'type']

filtered_df.loc[filtered_df['team'] != 'Bayer Leverkusen', 'type'] = 'ðŸ”µ' + \
filtered_df.loc[filtered_df['team'] != 'Bayer Leverkusen', 'type']

filtered_df['located_type'] = filtered_df['y'].astype(str) + ',' + \
    filtered_df['x'].astype(str) + ',' + filtered_df['type']
```

```python
actions_sequence = filtered_df.located_type.tolist()
match_ids = filtered_df.match_id.tolist()

action_mapping = {action: idx for idx, action in enumerate(set(actions_sequence))}
actions_numeric = [action_mapping[action] for action in actions_sequence]

num_actions = len(action_mapping)
transition_counts_by_match = {}

# Group by match_id
for match_id, group in filtered_df.groupby('match_id'):
    actions_sequence_match = group.located_type.tolist()
    actions_numeric_match = [action_mapping[action] for action in actions_sequence_match]
    
    # Initialize transition count matrix for each match
    transition_counts = np.zeros((num_actions, num_actions), dtype=int)
    
    # Count transitions for the current match
    for (current, next_) in zip(actions_numeric_match, actions_numeric_match[1:]):
        transition_counts[current, next_] += 1
    
    transition_counts_by_match[match_id] = transition_counts

# Now calculate the mean transition matrix
transition_matrix_sum = np.zeros((num_actions, num_actions), dtype=float)
num_matches = len(transition_counts_by_match)

for match_id, transition_counts in transition_counts_by_match.items():
    for i in range(num_actions):
        row_sum = np.sum(transition_counts[i])
        if row_sum > 0:
            transition_matrix_sum[i] += transition_counts[i] / row_sum

# Calculate the average transition matrix
transition_matrix_mean = transition_matrix_sum / num_matches

# Convert to DataFrame for easier inspection
transition_df_mean = pd.DataFrame(transition_matrix_mean, index=action_mapping.keys(), columns=action_mapping.keys())
transition_df = transition_df_mean.copy()
transition_df.sort_index()

# transition_df.to_csv('data/mean_transition_matrix.csv')
```

```python
def closest_points(y, x, pitch_width=16, pitch_height=24, num_points=5):
    points = [(i, j) for i in range(1, pitch_height+1) for j in range(1, pitch_width+1)]
    # Euclidean distance from (y, x) to each point
    distances = []
    for (y_prime, x_prime) in points:
        distance = np.sqrt((y_prime - y)**2 + (x_prime - x)**2)
        distances.append((distance, (y_prime, x_prime)))
    distances.sort(key=lambda item: item[0])
    closest = [point for _, point in distances[:num_points]]
    return closest
```

```python
def generate_match_sequence(transition_df, start_action, length, end_action=None):
    import numpy as np
    
    # Extract actions from the index and columns
    actions = transition_df.index.tolist()
    defending_actions = [
        'Interception', 
        'Ball Recovery', 
        'Dispossessed', 
        'Block', 
        'Clearance', 
        'Foul Committed', 
        'Goal Keeper',
        'Duel',
        'Pressure',
    ]
    
    # Initialize the sequence with the starting action
    sequence = [start_action]
    current_action = start_action
    last_attack_coordinates = None
    
    # Iterate to generate the sequence
    for _ in range(length - 1):
        current_idx = actions.index(current_action)  # Get index of the current action
        next_action_probs = transition_df.iloc[current_idx].values
        
        if np.sum(next_action_probs) != 0:
            next_action_probs = next_action_probs / np.sum(next_action_probs)
        else:
            next_action_probs = np.ones_like(next_action_probs) / len(next_action_probs)
        
        next_action = np.random.choice(actions, p=next_action_probs)
        
        # Handle defending actions based on the previous attacking action's location
        if any(def_action in next_action for def_action in defending_actions):
            _, _, action_type = next_action.split(',')
            if last_attack_coordinates:
                # Use the last attacking action's coordinates
                y, x = last_attack_coordinates
            else:
                # Default to current action's coordinates if no previous attack
                y, x = map(int, current_action.split(',')[:2])
            
            closest = closest_points(y, x)
            y, x = closest[0]
            next_action = f"{y},{x},{action_type}"
        else:
            # Update the last attack coordinates for non-defending actions
            y, x, _ = next_action.split(',')
            last_attack_coordinates = (int(y), int(x))
        
        sequence.append(next_action)
        current_action = next_action

        if end_action and current_action == end_action:
            break

    return sequence

# Example usage
example_sequence = generate_match_sequence(transition_df, start_action="12,9,ðŸ”´Pass", length=250)
for action in example_sequence:
    print(action)
```
