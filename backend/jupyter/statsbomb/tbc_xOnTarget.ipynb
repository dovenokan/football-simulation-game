```python
master_events_df_leverkusen = pd.read_csv("data/Bayer Leverkusen Bundesliga 23-24 Statsbomb Events Data.csv", low_memory=False)
# master_events_df = pd.read_csv("./data/master_events_processed.csv", low_memory=False)
master_events_df_ligue1 = pd.read_csv("./data/Ligue1 22-23 Statsbomb Events Data.csv", low_memory=False)
```

```python
master_events_df = pd.concat([master_events_df_leverkusen, master_events_df_leverkusen])
```

```python
def safe_literal_eval(val):
    try:
        return ast.literal_eval(val) if isinstance(val, str) and val != '[]' else []
    except (ValueError, SyntaxError):
        return []

master_events_df['location'] = master_events_df['location'].fillna('[]')
master_events_df['location'] = master_events_df['location'].apply(safe_literal_eval)
master_events_df['y'] = master_events_df['location'].apply(lambda loc: loc[0] if loc else None)
master_events_df['x'] = master_events_df['location'].apply(lambda loc: loc[1] if loc else None)
```

```python
# x on target
master_events_df.iloc[3574].to_dict()
```

```python
shot_df = master_events_df.copy().query("type == 'Shot'")
shot_df = shot_df.replace(['Saved to Post', 'Saved Off Target', 'Wayward' , 'Post'], 'Off T')
shot_df = shot_df.replace(['Saved'], 'Goal')
shot_df = shot_df.replace(['Goal'], 'On T')
shot_df["y"] = shot_df["y"].fillna(0).apply(lambda y: min(24, int(y * 24 / 120)))  
shot_df["x"] = shot_df["x"].fillna(0).apply(lambda x: min(16, int(x * 16 / 80)))   
shot_df["shot_on_target"] = shot_df["shot_outcome"].apply(lambda x: 1 if x == "On T" else 0)
```

```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
import numpy as np

df = shot_df.copy() 

# Features (x, y) and target (on_target)
X = df[['x', 'y']]
y = df['shot_on_target']

# Train logistic regression
model = LogisticRegression()
model.fit(X, y)

# Model coefficients
beta_0 = model.intercept_[0]
beta_1, beta_2 = model.coef_[0]

print(f"Intercept (β₀): {beta_0}")
print(f"Coefficient for x (β₁): {beta_1}")
print(f"Coefficient for y (β₂): {beta_2}")

# Function to calculate P(On T)
def expected_on_target(y, x):
    if y < 13:
        return 0
    beta_0 = -4.877
    beta_1 = -0.020
    beta_2 = 0.220
    z = beta_0 + beta_1 * x + beta_2 * y
    return 1 / (1 + np.exp(-z))

# Example prediction
for y in range(1,24+1):
    for x in range(1,16+1):
        shot_location = (y, x)
        p_on_target = expected_on_target(*shot_location)
        print(f"Expected probability of On T for location {shot_location}: {p_on_target:.2f}")
```
