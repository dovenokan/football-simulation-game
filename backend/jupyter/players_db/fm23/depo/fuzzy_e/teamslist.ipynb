```python
import pandas as pd
import numpy as np
import seaborn as sns
import re
```

```python
fifa_df = pd.read_csv("../fifa/pdb_23.csv")
fm_df = pd.read_csv("depo/fm2023_uluta.csv")#.sort_values(by="Values", ascending=False)
```

```python
# row values stripped.

for col in fm_df.select_dtypes("object").columns:
   fm_df[col] = fm_df[col].apply(lambda x: x.strip())
```

```python
# columns stripped.

fm_df.columns = [col.strip() for col in fm_df.columns]
```

```python
convert_date_format = lambda x: pd.to_datetime(x, format='%d/%m/%Y').strftime('%Y-%m-%d')
fm_df['DoB'] = fm_df['DoB'].apply(lambda x: x.split(" ")[0])
fm_df['DoB'] = fm_df['DoB'].apply(convert_date_format)
```

```python
fm_df['Height'] = fm_df['Height'].apply(lambda x: x.split(" ")[0])
```

```python
fifa_df["fifa_fuzzy_name"] = fifa_df["dob"] + " " + fifa_df["height_cm"].astype(str) 
fm_df["fm_fuzzy_name"] = fm_df["DoB"] + " " + fm_df["Height"].astype(str) 
```
