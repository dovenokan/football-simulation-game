<h1 style="color:aqua;background:black;"> Positions tpr Weights</h1>

```python
plat_df = pd.read_csv("players_db/fm23/platinum22_fm23data.csv")
inside_df = pd.read_csv("players_db/fm23/fm22_players_fminside.csv")
```

```python
tpr_cols = ['Aggression', 'Anticipation', 'Bravery', 'Composure', 'Concentration',
       'Decisions', 'Determination', 'Flair', 'Leadership', 'Off the Ball',
       'Positioning', 'Teamwork', 'Vision', 'Work Rate', 'Acceleration',
       'Agility', 'Balance', 'Jumping Reach', 'Natural Fitness', 'Pace', 'Stamina',
       'Strength', 'Corners', 'Crossing', 'Dribbling', 'Finishing',
       'First Touch', 'Free Kick Taking', 'Heading', 'Long Shots',
       'Long Throws', 'Marking', 'Passing', 'Penalty Taking', 'Tackling',
       'Technique']
len(tpr_cols)
```

```python
plat_df.columns
```

```python
new_column_names = {
    'Aggressiion': 'Aggression',
    'Decision': 'Decisions',
    'Off The Ball': 'Off the Ball',
    'Position.1': 'Positioning',
    'Natural Fitness': 'Natural Fitness',
}
plat_df.rename(columns=new_column_names, inplace=True)
```

```python
positions = ['DL', 'DC', 'DR', 'WBL', 'WBR', 'DM', 'ML', 'MC', 'MR', 'AML', 'AMC', 'AMR', 'ST'] # no gk
plat_df['Best_Position'] = plat_df[positions].idxmax(axis=1)
```

```python
inside_df['Best_Position'] = inside_df['Position'].apply(lambda x: x.split(", ")[0] if isinstance(x, str) else [])
inside_df.columns = [col.replace(" ", "_") for col in inside_df.columns]
plat_df.columns = [col.replace(" ", "_") for col in plat_df.columns]
tpr_cols = [col.replace(" ", "_") for col in tpr_cols]
```

```python
formula_df = pd.DataFrame(index=tpr_cols)
inside_df['tpr'] = None
plat_df = inside_df.sort_values(by="Ability", ascending=False)[:200]

for pos in positions:
    query = plat_df['Best_Position'] == pos
    top_50_players = plat_df[query].head(20)
    
    mean_values = (top_50_players[tpr_cols].sum() / top_50_players[tpr_cols].count())
    
    percentage_values = mean_values / mean_values.sum()
    sorted_percentages = percentage_values.sort_values()
    formula_df[pos] = sorted_percentages.to_dict()

    plat_df.loc[query, 'tpr'] = (plat_df[query][tpr_cols] * percentage_values).sum(axis=1)

    max_tpr = plat_df.loc[query, 'tpr'].max()
    min_tpr = plat_df.loc[query, 'tpr'].min()
    if max_tpr == min_tpr:
        min_tpr -= 1
        plat_df.loc[query, 'tpr'] = (plat_df.loc[query, 'tpr'] - min_tpr) / (max_tpr - min_tpr)

    multiplier_tpr = (94 / plat_df.loc[query, 'tpr'].max())
    plat_df.loc[query, 'tpr'] = (plat_df.loc[query, 'tpr'] * multiplier_tpr).astype(int)

plat_df["Error"] = plat_df["tpr"] - plat_df["Ability"]
plat_df["Error"].mean()
```

```python
plat_df.sort_values(by="tpr", ascending=False)[["Name","Ability","tpr","Best_Position","Error"]].head(n=5)
```

```python
formula_df["GK"] = 0
gk_attributes = [ "Aerial_Reach","Command_of_Area", "Communication","Eccentricity", "First_Touch", "Handling", "Kicking", "One_on_Ones", "Punching", "Reflexes", "Rushing_Out", "Throwing", "Passing" ]
for i in gk_attributes:
    formula_df.loc[i, "GK"] = 1 / len(gk_attributes)
formula_df = formula_df.fillna(0)
```

```python
formula_df.sum()
```

```python
formula_df["ST"].sort_values(ascending=False) # 11 tane gk-only attributes
```

```python
# Shifting Positions positions=non_gk_attributes
for pos in positions:
    bottom_7 = formula_df[pos].sort_values(ascending=False)[-18:-11]
    top_7 = formula_df[pos].sort_values(ascending=False)[0:7]
    formula_df.loc[bottom_7.index, pos] = 0
    formula_df.loc[top_7.index, pos] = list(np.array(bottom_7) + np.array(top_7))
```
