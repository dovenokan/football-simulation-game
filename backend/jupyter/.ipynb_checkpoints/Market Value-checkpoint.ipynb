```python
fm_df = pd.read_csv("players_db/fm23/fm23db_processed.csv")
team_df = pd.read_csv("players_db/fm23/team_ratings.csv")[["Club","tpr"]]
```

<h2 style="color:aqua;">  Hands on "Market_Value"</h2>

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OrdinalEncoder
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Load the data

data = fm_df.copy().dropna()

# Preprocess the data
encoder = OrdinalEncoder()
data['Best_Pos_Encoded'] = encoder.fit_transform(data[['Best_Pos']])

# Define the feature set and target variable
X = data[["Wage",'tpr', 'Age', 'Best_Pos_Encoded','tpr_team']]
y = data['Market_Value']

# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

# Train the model
model = LinearRegression()
model.fit(X_train, y_train)

# Evaluate the model
y_pred = model.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f'MSE: {mse}')
print(f'MAE: {mae}')
print(f'R-squared: {r2}')

# Predict market values for new data
def Predict_Market_Value(model, wage, tpr, age, position, tpr_team):
    new_data = pd.DataFrame({
        'Wage': [wage],
        'tpr': [tpr],
        'Age': [age],
        'Best_Pos_Encoded': [encoder.transform([[position]])[0][0]],
        'tpr_team': [tpr_team],
    })
    predicted_market_value = model.predict(new_data)[0]
    output = max(5000, predicted_market_value)
    return int(output)
```

```python
t_wage = 100
t_tpr = 92
t_age = 24
t_pos = [encoder.transform([["AMR"]])[0][0]][0]
t_team_tpr = 66
sum([
    model.coef_[0]*t_wage,
    model.coef_[1]*t_tpr,
    model.coef_[2]*t_age,
    model.coef_[3]*t_pos,
    model.coef_[4]*t_team_tpr,
    model.intercept_
])
```

```python
fm_df["Predicted_Market_Value"] = fm_df.apply(lambda row: \
                                    Predict_Market_Value(model, row["Wage"],row["tpr"],row["Age"],row["Best_Pos"],row["tpr_team"]), axis=1)

fm_df["Market_Value_Million"] = (fm_df["Predicted_Market_Value"] /  10**6).astype(int)
```

```python
fm_df["Market_Value"] = fm_df["Predicted_Market_Value"]
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OrdinalEncoder
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Load the data

data = fm_df.copy().dropna()

# Preprocess the data
encoder = OrdinalEncoder()
data['Best_Pos_Encoded'] = encoder.fit_transform(data[['Best_Pos']])

# Define the feature set and target variable
X = data[["Market_Value", 'tpr', 'Age', 'Best_Pos_Encoded','tpr_team']]
y = data['Wage']

# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

# Train the model
model = LinearRegression()
model.fit(X_train, y_train)

# Evaluate the model
y_pred = model.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f'MSE: {mse}')
print(f'MAE: {mae}')
print(f'R-squared: {r2}')

# Predict market values for new data
def Predict_Market_Value(model, market_value, tpr, age, position, tpr_team):
    new_data = pd.DataFrame({
        'Market_Value': [market_value],
        'tpr': [tpr],
        'Age': [age],
        'Best_Pos_Encoded': [encoder.transform([[position]])[0][0]],
        'tpr_team': [tpr_team],
    })
    predicted_market_value = model.predict(new_data)[0]
    output = max(100, predicted_market_value)
    return int(output)
```

```python
t_wage = 100
t_tpr = 92
t_age = 24
t_pos = [encoder.transform([["AMR"]])[0][0]][0]
t_team_tpr = 66
sum([
    model.coef_[0]*t_wage,
    model.coef_[1]*t_tpr,
    model.coef_[2]*t_age,
    model.coef_[3]*t_pos,
    model.coef_[4]*t_team_tpr,
    model.intercept_
])
```

```python
fm_df["Predicted_Wage"] = fm_df.apply(lambda row: \
                          Predict_Market_Value(model, row["Market_Value"],row["tpr"],row["Age"],row["Best_Pos"],row["tpr_team"]), axis=1)

fm_df["Wage_K"] = (fm_df["Predicted_Wage"] /  10**3).astype(int)
```

```python
def Normalize(X, min_val, max_val):
    min_col = X.min()
    max_col = X.max()
    value = (X - min_col) / (max_col - min_col) * (max_val - min_val) + min_val
    return value.astype(int)

fm_df["Raw_Predicted_Wage"] = fm_df["Predicted_Wage"]
fm_df["Predicted_Wage"] = Normalize(fm_df["Raw_Predicted_Wage"], 100, 1.5*10**6)
```
